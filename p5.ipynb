```python
import pandas as pd
```

```python
df = pd.read_csv('bitcoin.csv')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Timestamp'], unit='s')
df.head()
```

```python
df[df['Close'] > 50]
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
```

```python
df.groupby('Year')['Close'].mean()
```

```python
df.groupby('Month')['Close'].mean()
```

```python
#Aggregate Functions
df['Close'].mean()
```

```python
df['High'].max()
```

```python
df['Low'].min()
```

```python
df['Volume'].sum()
```

```python
df['Close'].count()
```

```python
df.groupby('Year').agg({
    'Close': ['mean', 'min', 'max'],
    'Volume': ['sum', 'mean']
})
```

```python
df.loc[df['High'].idxmax()]
```

```python
df.loc[df['Low'].idxmin()]
```

```python
import matplotlib.pyplot as plt

plt.plot(df['Date'], df['Close'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Bitcoin Price Trend')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot(df['Date'], df['Volume'])
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Bitcoin Trading Volume Over Time')
plt.show()
```
