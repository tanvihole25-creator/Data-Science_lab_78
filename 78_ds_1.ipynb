```python
def display_multiplication_table(number):
    print(f"Multiplication Table for {number}:")
    for i in range(1, 11): # Table from 1 to 10
        print(f"{number} x {i} = {number * i}")

# Example usage:
num_to_display = 7
display_multiplication_table(num_to_display)

print("\n") # Add a newline for better readability between examples

num_to_display = 12
display_multiplication_table(num_to_display)
```

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

# Example usage:
num = 5
print(f"The factorial of {num} is {factorial(num)}")

num = 0
print(f"The factorial of {num} is {factorial(num)}")

num = -3
print(f"The factorial of {num} is {factorial(num)}")
```

def is_palindrome(number):
    # Convert the number to a string to easily reverse it
    num_str = str(number)
    # Reverse the string
    reversed_num_str = num_str[::-1]

    # Compare the original and reversed strings
    if num_str == reversed_num_str:
        return f"{number} is a palindrome."
    else:
        return f"{number} is not a palindrome."

# Example usage:
num1 = 121

# Task
Develop an expert system by defining its rules, implementing the logic in Python, and demonstrating its decision-making capabilities with examples. Finally, summarize the purpose and functionality of the developed expert system.

## Define Expert System Rules

### Subtask:
Establish a set of simple rules for the expert system's decision-making process. This will involve choosing a domain and defining 'if-then' conditions.


For this expert system, I will choose the domain of **Animal Identification**.

Here are the simple 'if-then' rules that will form the basis of our expert system:

1.  **Rule 1: Bird Identification**
    *   **IF** an animal has feathers **AND** lays eggs
    *   **THEN** it is a bird

2.  **Rule 2: Mammal Identification**
    *   **IF** an animal has fur/hair **AND** gives birth to live young **AND** produces milk
    *   **THEN** it is a mammal

3.  **Rule 3: Fish Identification**
    *   **IF** an animal lives in water **AND** has gills **AND** has scales
    *   **THEN** it is a fish

4.  **Rule 4: Reptile Identification**
    *   **IF** an animal has scales **AND** lays eggs **AND** is cold-blooded
    *   **THEN** it is a reptile

## Implement Expert System Logic

### Subtask:
Write Python code to implement the expert system. This will include a function that takes user input, applies the defined rules, and provides a decision.


**Reasoning**:
I need to write a Python function that implements the expert system logic to identify animals based on given characteristics, using conditional statements as per the instructions.



```python
def identify_animal(
    has_feathers=False,
    lays_eggs=False,
    has_fur_hair=False,
    gives_live_birth=False,
    produces_milk=False,
    lives_in_water=False,
    has_gills=False,
    has_scales=False,
    is_cold_blooded=False
):
    """
    Identifies an animal based on provided characteristics using expert system rules.
    """
    # Rule 1: Bird
    if has_feathers and lays_eggs:
        return "Bird"
    # Rule 2: Mammal
    elif has_fur_hair and gives_live_birth and produces_milk:
        return "Mammal"
    # Rule 3: Fish
    elif lives_in_water and has_gills and lays_eggs and is_cold_blooded:
        return "Fish"
    # Rule 4: Reptile
    elif has_scales and lays_eggs and is_cold_blooded and not lives_in_water:
        return "Reptile"
    # Rule 5: Amphibian (example rule, common characteristics)
    elif lays_eggs and lives_in_water and is_cold_blooded and not has_gills:
        return "Amphibian"
    # Default: Unknown
    else:
        return "Unknown animal type based on current rules."

# Example usage of the function:
print("--- Animal Identification Examples ---")

# Example 1: Bird
print(f"Characteristics: has_feathers=True, lays_eggs=True")
print(f"Identified: {identify_animal(has_feathers=True, lays_eggs=True)}\n")

# Example 2: Mammal
print(f"Characteristics: has_fur_hair=True, gives_live_birth=True, produces_milk=True")
print(f"Identified: {identify_animal(has_fur_hair=True, gives_live_birth=True, produces_milk=True)}\n")

# Example 3: Fish
print(f"Characteristics: lives_in_water=True, has_gills=True, lays_eggs=True, is_cold_blooded=True")
print(f"Identified: {identify_animal(lives_in_water=True, has_gills=True, lays_eggs=True, is_cold_blooded=True)}\n")

# Example 4: Reptile (e.g., snake, lizard)
print(f"Characteristics: has_scales=True, lays_eggs=True, is_cold_blooded=True, not lives_in_water")
print(f"Identified: {identify_animal(has_scales=True, lays_eggs=True, is_cold_blooded=True, lives_in_water=False)}\n")

# Example 5: Amphibian (e.g., frog)
print(f"Characteristics: lays_eggs=True, lives_in_water=True, is_cold_blooded=True, not has_gills")
print(f"Identified: {identify_animal(lays_eggs=True, lives_in_water=True, is_cold_blooded=True, has_gills=False)}\n")

# Example 6: Unknown animal
print(f"Characteristics: has_fur_hair=True, lays_eggs=True")
print(f"Identified: {identify_animal(has_fur_hair=True, lays_eggs=True)}\n")

# Example 7: Only lives in water and cold blooded
print(f"Characteristics: lives_in_water=True, is_cold_blooded=True")
print(f"Identified: {identify_animal(lives_in_water=True, is_cold_blooded=True)}\n")
```
