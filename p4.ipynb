```python

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# User input
n = int(input("Enter number of values: "))
ch = input("Enter any character: ")

# Input values
print("Enter the numbers:")
data = []
for i in range(n):
    data.append(int(input()))

data = np.array(data)

# Mean, Median, Mode
mean_val = np.mean(data)
median_val = np.median(data)
mode_val = stats.mode(data, keepdims=True)[0][0]

# Output
print("\nYou entered character:", ch)
print("Mean   :", mean_val)
print("Median :", median_val)
print("Mode   :", mode_val)

# Graph
plt.hist(data)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Data Distribution")
plt.show()
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Take integer input from user
data = list(map(int, input("Enter integer data separated by space: ").split()))

# Convert to NumPy array
arr = np.array(data)

# Statistical calculations
mean_val = np.mean(arr)
median_val = np.median(arr)
variance_val = np.var(arr)
std_dev_val = np.std(arr)

# SciPy example (mode)
mode_val = stats.mode(arr, keepdims=True)

# Display results
print("\nStatistical Parameters:")
print("Mean =", mean_val)
print("Median =", median_val)
print("Variance =", variance_val)
print("Standard Deviation =", std_dev_val)
print("Mode =", mode_val.mode[0])

# Plotting using Matplotlib
plt.figure()
plt.hist(arr)
plt.title("Histogram of Data")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
# Plot histogram with better visibility
plt.figure(figsize=(8, 6))          # Increase figure size
plt.hist(arr, bins=10, edgecolor='black')

plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Dataset Distribution")

plt.grid(True)                       # Add grid for clarity
plt.show()
```

```python
# Number of coin tosses
tosses = int(input("Enter number of coin tosses: "))

# Simulate coin toss (0 = Head, 1 = Tail)
results = np.random.randint(0, 2, tosses)

# Count outcomes
heads = np.sum(results == 0)
tails = np.sum(results == 1)

# Calculate probabilities
print("\nCoin Toss Simulation Results:")
print("Heads:", heads)
print("Tails:", tails)
print("Probability of Head:", heads / tosses)
print("Probability of Tail:", tails / tosses)
```
